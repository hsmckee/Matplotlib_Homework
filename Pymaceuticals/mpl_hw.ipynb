```python
#Dependencies
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt


```

```python
#read csv and creat datafram
trial_df = pd.read_csv(os.path.join("data", "clinicaltrial_data.csv"))
drug_df = pd.read_csv(os.path.join("data", "mouse_drug_data.csv"))
```

```python
#Clean up data
#change column names to usable names
trial_df.columns = trial_df.columns.str.replace(' ','_')
drug_df.columns = drug_df.columns.str.replace(' ', '_')
```

```python
# Combine the data into a single dataset
total = pd.merge(trial_df, drug_df, on='Mouse_ID' )
print (total)
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
#q = total.groupby(['Drug','Timepoint']).mean()
q = total.groupby(['Drug','Timepoint']).mean().reset_index()

```

```python
#Creating a scatter plot that shows how the tumor volume changes over time for each treatment.
#fig = plt.scatter(q.'Timepoint', q.'Tumor_Volume_(mm3)')
#q.plot(kind='scatter', x='Timepoint', y='Tumor_Volume_(mm3)')

drug_colors = {'Capomulin':'red', 'Infubinol':'blue', 'Ketapril':'yellow', 'Placebo':'green', 'Ceftamin':'orange', 'Naftisol':'teal', 'Propriva':'violet',
              'Ramicane':'lime', 'Stelasyn':'grey', 'Zoniferol':'olive'}

plt.scatter('Timepoint', 'Tumor_Volume_(mm3)', data=q, c=q['Drug'].apply(lambda x: drug_colors[x]))
plt.title ('Tumor Volume Change Over Time for Drug Taken')
plt.xlabel ('Time (days)')
plt.ylabel ('Tumor Volume (mm3)')
plt.legend (list(drug_colors.keys()))

plt.show
```

```python
#Creating a scatter plot that shows how the number of metastatic sites changes over time for each treatment.
plt.scatter('Timepoint', 'Metastatic_Sites', data=q, c=q['Drug'].apply(lambda x: drug_colors[x]))
plt.title ('Metastatic Site Change Over Time for Drug Taken')
plt.xlabel ('Time (days)')
plt.ylabel ('Metastatic Sites (ct)')
plt.legend (list(drug_colors.keys()))

plt.show
```

```python
#Creating a scatter plot that shows the number of mice still alive through the course of treatment (Survival Rate)
#survival = total.groupby(['Drug','Timepoint'], as_index=False).nunique('Mouse_ID')
survival = total.groupby(['Drug','Timepoint']).agg(['nunique']).reset_index()
#survival = survival.reset_index(drop=True)
#survival.get_group('Mouse_ID', asindex).reset_index()
print (survival)


plt.scatter('Timepoint', 'Mouse_ID', data=survival, c=q['Drug'].apply(lambda x: drug_colors[x]))
plt.title ('Survival Rate')
plt.xlabel ('Time (days)')
plt.ylabel ('Number of Mice Alive')
plt.legend (list(drug_colors.keys()))

plt.show
```

```python
#Creating a bar graph that compares the total % tumor volume change for each drug across the full 45 days.
#tumor_volume_change = total.groupby('Drug', sort=False).apply(lambda x: (x['Tumor_Volume_(mm3)'].values.reset_index(name='pct_vol_change')))
total1 = total.set_index('Timepoint')
print (total1.index)
timepoint0 = total1.loc[[0]]
timepoint45 = total1.loc[[45]]
tumor_volume_change0 = timepoint0.groupby(['Drug']).mean().reset_index()
tumor_volume_change45 = timepoint45.groupby(['Drug']).mean().reset_index()
#print (timepoint0)
#tumor_volume_change = (tumor_volume_change45['Tumor_Volume_(mm3)'] - tumor_volume_change0['Tumor_Volume_(mm3)'])/tumor_volume_change45['Tumor_Volume_(mm3)']*100
#tvc = pd.merge(timepoint0, tumor_volume_change, on='' )

results = tumor_volume_change0[['Drug']]
results['Percent'] = (tumor_volume_change45['Tumor_Volume_(mm3)'] - tumor_volume_change0['Tumor_Volume_(mm3)'])/tumor_volume_change45['Tumor_Volume_(mm3)']*100
print(results)


#print (tumor_volume_change)
#print(tumor_volume_change0)
plt.bar('Drug', 'Percent', data=results, color = results['Drug'].apply(lambda x: drug_colors[x]))
plt.title ('% Tumor Volume Change')
plt.xlabel ('Time (days)')
plt.ylabel ('% Volume')
plt.xticks(rotation = 70)
plt.show
```
